```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

```

```python
df = pd.read_csv(r"C:\Users\joe\Downloads\Loan.csv")
```

```python
df.head()
```

# DATA CLEANING

```python
df.isnull()
```

```python
total = df.isnull().sum() #getting count for number of null values in each column
total
```

```python
df['Gender'].value_counts()  #finding the Male/Female split in the data in otrder to fill the mising values
```

```python
df.Gender = df.Gender.fillna('Male') #filling missing values with value=Male as no of males is 4times more 
```

```python
df['Married'].value_counts() #finding married data split 
```

```python
df.Married = df.Married.fillna('Yes') #filling missing values as "Yes" as majority data is Yes 
```

```python
df['Dependents'].value_counts() #finding the split for the dependent data 
```

```python
df.Dependents = df.Dependents.fillna('0') #filling missing values with '0' as '0' has majority 
```

```python
df['Self_Employed'].value_counts() #finding the split of data here 
```

```python
df.Self_Employed = df.Self_Employed.fillna('No') #filling missing data with No
```

```python
df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.mean()) #filling the missing values with the mean of all the data 
```

```python
df['Loan_Amount_Term'].value_counts() #finding the split of data
```

```python
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(360.0) #filling missing data with 360
```

```python
df['Credit_History'].value_counts() #finding the data split 
```

```python
df.Credit_History = df.Credit_History.fillna(1.0) #filling missing values with the majority 
```

```python
total = df.isnull().sum() #getting count for number of null values in each column with all the updated parts
total
```

```python
df.head(30)
```

# Converting Categorical Varibale into Numeric


```python
from sklearn.preprocessing import LabelEncoder 
var_mod = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
for i in var_mod:
    df[i]=le.fit_transform(df[i])  #converting all the non numerical data into machine readable numeric data 
```

```python
df.head(30)
```

# Feature Selection

```python
X=df.iloc[:,1:12] #independent variable
Y=df.iloc[:,12]   #dependent variable
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
#splitting the data into training and test set in the ratio of 8:2
```

```python
x_train
```

```python
y_train
```

# Building Decision Tree Model 

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_prediction = model.predict(x_test)
```

```python
y_prediction
```

```python
y_test
```

# Model Evaluation of Decision Tree

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,y_prediction))
#accuracy is lower than ideal and changing according to the tain test split 
```

```python
cnf_matrix=metrics.confusion_matrix(y_test,y_prediction)
cnf_matrix
```

# Visualization of Confusion Matrix for Decesion Tree

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sn.heatmap(cnf_matrix ,annot=True)
plt.title("Confusion Matrix")
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
plt.show()
```

# Buliding Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_regression=LogisticRegression()
logistic_regression.fit(x_train,y_train)
```

```python
y_pred=logistic_regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

# Model Evaluation For Logistic Regression

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(y_test,y_pred)
print (matrix)
```
